```python
import numpy as np
from IPython.display import display
import pandas as pd
```

```python
products_df = pd.DataFrame({
'Product_ID': [1, 2, 3],
'Product_Name': ['iPhone 12', 'Samsung S21', 'Pixel 5'],
'Product_Price': [999, 899, 699],
})

sales_df = pd.DataFrame({
'Sale_ID': [1, 2, 3, 4, 5, 6, 7],
'Sale_Date': pd.to_datetime(['2021-01-01', '2021-02-01', '2021-02-15', '2020-03-01', '2020-04-01',
'2020-02-15', '2020-01-01']),
'Customer_ID': [1, 2, 3, 2, 3, 4, 4],
'Product_ID': [1, 1, 2, 2, 3, 3, 1],
})

customers_df = pd.DataFrame({
'Customer_ID': [1, 2, 3, 4],
'Customer_Name': ['John Doe', 'Jane Doe', 'Jim Brown', 'Jake Smith'],
'Customer_Age': [25, 30, 35, 20],
'Customer_Email': ['john@doe.com', 'jane@doe.com', 'jim@brown.com', 'jake@smith.com'],
})
```

```python
display(products_df)
products_df = products_df.copy()
display(products_df.info())
display(products_df.describe().T)
```

```python
display(sales_df)
sales_dff = sales_df.copy()
display(sales_df.info())
display(sales_df.describe().T)
```

```python
display(customers_df)
customers_df = customers_df.copy()
display(customers_df.info())
display(customers_df.describe().T)
```

```python
# 1.

merged_df = pd.merge(sales_df, products_df, on='Product_ID', how='inner')
merged_df = pd.merge(merged_df, customers_df, on='Customer_ID', how='inner')

unique_customers = merged_df[['Product_Name', 'Customer_ID', 'Customer_Age']].drop_duplicates()

result = unique_customers.groupby('Product_Name').agg({
    'Customer_ID': 'count',
    'Customer_Age': 'mean'
}).rename(columns={
    'Customer_ID': 'Unique_Customers',
    'Customer_Age': 'Avg_Age'
}).round(1).reset_index()
result
```

```python
# 2.

merged_df = pd.merge(sales_df, products_df, on='Product_ID', how='inner')

merged_df['Sale_Year'] = merged_df['Sale_Date'].dt.year

sales_product_per_year = merged_df.groupby(['Sale_Year', 'Product_Name']).agg({
    'Sale_ID': 'count'
}).rename(columns={'Sale_ID': 'Sales_Count'}).reset_index()
sales_product_per_year
```
