```python
import numpy as np
from IPython.display import display
import pandas as pd
```

```python
sales_funnel_CRM = pd.read_csv("Sales_Funnel_CRM.csv")
display(sales_funnel_CRM)
df_sales_funnel_CRM = sales_funnel_CRM.copy()
display(df_sales_funnel_CRM.info())
display(df_sales_funnel_CRM.describe().T)
```

```python
# 1.

display(df_sales_funnel_CRM['Product'].unique())
print()

pivot_df = df_sales_funnel_CRM.pivot_table(
    values='Licenses',        
    index='Company',          
    columns='Product',        
    aggfunc='sum',                      
)

pivot_df
```

```python
# 2.

df_totals = df_sales_funnel_CRM.groupby('Company').agg({
    'Sale Price': 'sum',
    'Licenses': 'sum'
}).rename(columns={
    'Sale Price': 'total_sale_price',
    'Licenses': 'total_licenses'
}).reset_index()

df_totals['total_value'] = df_totals['total_sale_price'] + df_totals['total_licenses']

df_totals
```
